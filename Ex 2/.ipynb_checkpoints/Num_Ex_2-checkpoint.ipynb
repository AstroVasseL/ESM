```python
import numpy as np
import matplotlib.pyplot as plt

# Define the grid parameters
dx = 25e3   # Spatial step size
dt = 2000  # Temporal step size
L = 2500e3   # Domain length
T = 500*3600     # Simulation time
c0 = 10    # Constant value for initial condition

# Define the velocity and initial condition
u = 10 #m/s

def initial_condition(x):
    if x < 1125e3 or x > 1375e3:
        return 0
    else:
        return c0

# Create the grid
x = np.arange(0, L, dx)
t = np.arange(0, T, dt)

# Initialize the solution matrix
C = np.zeros((len(t), len(x)))

# Set the initial condition
for j in range(len(x)):
    C[0,j] = initial_condition(x[j])

# Compute the solution using the Euler forward in time and upwind in space method
for i in range(len(t)-1):
    for j in range(1, len(x)):
        C[i+1,j] = C[i,j] - (u*dt/dx)*(C[i,j]-C[i,j-1])
        # Apply periodic boundary conditions
    C[i+1,0] = C[i+1,-2]
    C[i+1,-1] = C[i+1,1]

# Plot the solution
plt.figure(figsize=(24,10))
plt.imshow(C, aspect='auto', cmap='coolwarm', origin='lower', extent=[0, L/1e3, 0, T/3600])

plt.colorbar()
plt.xlabel('X[km]')
plt.ylabel('t [h]')
plt.title('Numerical Solution to the Linear Advection Equation using the Euler-Forward scheme')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the grid parameters
dx = 25e3   # Spatial step size
dt = 2000  # Temporal step size
L = 2500e3   # Domain length
T = 200*3600     # Simulation time
c0 = 10    # Constant value for initial condition

# Define the velocity and initial condition
u0 = 10 #m/s

def initial_condition(x):
    if x < 1125e3 or x > 1375e3:
        return 0
    else:
        return c0

# Create the grid
x = np.arange(0, L, dx)
t = np.arange(0, T, dt)

# Initialize the solution matrix
C = np.zeros((len(t), len(x)))

# Set the initial condition
for j in range(len(x)):
    C[0,j] = initial_condition(x[j])

# Compute the solution using the Euler forward in time and upwind in space method
for i in range(len(t)-1):
    for j in range(1, len(x)-2):
        C[i+1,j] = C[i,j]-(u0*dt/(2*dx)) * ((C[i,j+1]-C[i,j-1])) + 0.5*(np.square(u0)*np.square(dt)/np.square(dx))*(C[i,j+1]-2*C[i,j]+C[i,j-1]) 
#         Apply periodic boundary conditions
    C[i+1,0] = C[i+1,-3]
    C[i+1,-2] = C[i+1,1]
    


# Plot the solution
plt.figure(figsize=(24,10))
plt.imshow(C, aspect='auto', cmap='coolwarm', origin='lower', extent=[0, L/1e3, 0, T/3600])

plt.colorbar()
plt.xlabel('X[km]')
plt.ylabel('t [h]')
plt.title('Numerical Solution to the Linear Advection Equation using Lax-Wendroff Scheme')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
c0 = 1.0       # wave speed
L = 25000.0     # domain length
nx = 2500       # number of grid points
dx = L/nx      # spatial step size
dt = 300.0      # time step
tmax = 3600*30.0# maximum simulation time

# Grid
x = np.linspace(0, L, nx+1)[:-1]  # spatial grid
k = np.fft.fftfreq(nx, d=dx)     # wavenumbers

# Initial condition
def initial_condition(x):
    u = np.zeros_like(x)
    u[(x >= 1125.0) & (x <= 1375.0)] = c0
    return u
u = initial_condition(x)

# Precompute spectral coefficients
expikdx = np.exp(-1j*k*c0*dt)
D = -1j*k*c0

# Time stepping
t = 0.0
while t < tmax:
    # Matsuno scheme: predictor step
    u_hat = np.fft.fft(u)
    u_hat_pred = expikdx * u_hat
    u_pred = np.real(np.fft.ifft(u_hat_pred))

    # Matsuno scheme: corrector step
    u_hat_pred = np.fft.fft(u_pred)
    u_hat_corr = (3*expikdx - expikdx**3) / 2 * u_hat - (expikdx + expikdx**3) / 2 * u_hat_pred
    u = np.real(np.fft.ifft(u_hat_corr))

    # Apply periodic boundary conditions
    u[0] = u[-2]
    u[-1] = u[1]

    # Update time
    t += dt

# Plot final solution
plt.plot(x, u, 'b-')
plt.xlabel('x')
plt.ylabel('u')
plt.show()
```
